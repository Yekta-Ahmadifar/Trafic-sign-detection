```python
import numpy as np
import tensorflow as tf
import pandas as pd

df_labels = pd.read_csv('labels.csv')
labele_map = dict(zip(df_labels['ClassId'], df_labels['Name']))


def predict_new_image(image_path, model_path='TrafficModel.keras'):
    # Load the saved model
    trained_model = tf.keras.models.load_model(model_path)
    
    # Load and preprocess the single image
    img = tf.keras.utils.load_img(image_path, target_size=(64, 64))
    img_array = tf.keras.utils.img_to_array(img)
    img_array = tf.expand_dims(img_array, 0) # Create a batch of 1
    
    # Predict
    predictions = trained_model.predict(img_array)
    predicted_class_id = np.argmax(predictions[0])
    confidence = np.max(predictions[0]) * 100
    
    # Get human-readable name from your map
    sign_name = labele_map.get(predicted_class_id, "Unknown")
    
    print(f"Result: {sign_name} ({confidence:.2f}% confidence)")

# Example usage:
predict_new_image('TEST/2/002_1_0022.png')
```
